a. Untuk masalah ini, saya akan memilih bar chart (diagram batang). Bar chart cocok untuk menunjukkan perbandingan antara berbagai kategori. Dalam hal ini, kita dapat menggunakan bar chart untuk menunjukkan seberapa sering masing-masing makanan penutup terjual dan dengan mudah melihat perbandingan popularitas relatifnya.

```python
import matplotlib.pyplot as plt

# Data
dessert_sales = {'Lava Cake': 14, 'Mousse': 5, 'Chocolate Cake': 12, 'Ice Cream': 19, 'Truffles': 6,
                'Brownie': 8, 'Chocolate Chip Cookie': 12, 'Chocolate Pudding': 9, 'Souffle': 10,
                'Chocolate Cheesecake': 17, 'Chocolate Chips': 2, 'Fudge': 9, 'Mochi': 13}

# Membuat bar chart
plt.figure(figsize=(10, 6))
plt.bar(dessert_sales.keys(), dessert_sales.values(), color='skyblue')
plt.xlabel('Makanan Penutup')
plt.ylabel('Penjualan Mingguan')
plt.title('Popularitas Makanan Penutup')
plt.xticks(rotation=45, ha='right')
plt.show()
```

a. Untuk masalah ini, saya akan memilih heatmap. Heatmap adalah grafik yang sangat cocok untuk menunjukkan pola atau perbandingan dalam data berbasis matriks seperti penggunaan CPU per jam selama seminggu. Heatmap dapat memberikan visualisasi yang jelas tentang kapan penggunaan CPU tinggi atau rendah.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Data
cpu_usage = [
    [2, 2, 4, 2, 4, 1, 1, 4, 4, 12, 22, 23, 45, 9, 33, 56, 23, 40, 21, 6, 6, 2, 2, 3],  # Monday
    [1, 2, 3, 2, 3, 2, 3, 2, 7, 22, 45, 44, 33, 9, 23, 19, 33, 56, 12, 2, 3, 1, 2, 2],  # Tuesday
    [2, 3, 1, 2, 4, 4, 2, 2, 1, 2, 5, 31, 54, 7, 6, 34, 68, 34, 49, 6, 6, 2, 2, 3],  # Wednesday
    [1, 2, 3, 2, 4, 1, 2, 4, 1, 17, 24, 18, 41, 3, 44, 42, 12, 36, 41, 2, 2, 4, 2, 4],  # Thursday
    [4, 1, 2, 2, 3, 2, 5, 1, 2, 12, 33, 27, 43, 8, 38, 53, 29, 45, 39, 3, 1, 1, 3, 4],  # Friday
    [2, 3, 1, 2, 2, 5, 2, 8, 4, 2, 3, 1, 5, 1, 2, 3, 2, 6, 1, 2, 2, 1, 4, 3],  # Saturday
    [1, 2, 3, 1, 1, 3, 4, 2, 3, 1, 2, 2, 5, 3, 2, 1, 4, 2, 45, 26, 33, 2, 2, 1],  # Sunday
]

# Membuat heatmap
sns.set()
plt.figure(figsize=(12, 8))
days = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
hours = np.arange(24)
sns.heatmap(cpu_usage, cmap="YlGnBu", xticklabels=hours, yticklabels=days, annot=True, fmt="d")
plt.xlabel('Jam (24 jam)')
plt.ylabel('Hari')
plt.title('Penggunaan CPU per Jam Selama Seminggu')
plt.show()
```

b. kita bisa melihat pola untuk menentukan jam makan siang dengan melihat warna yang lebih cerah pada setiap baris (hari). Pada umumnya, jam makan siang mungkin terjadi di sekitar jam 12:00-13:00. Dari data diatas pekerja mungkin makan siang pada jam 13.00

c. Dari data yang diberikan, kita dapat melihat bahwa penggunaan CPU pada hari Sabtu dan Minggu juga terlihat, terutama pada jam-jam tertentu. Oleh karena itu, kita dapat menyimpulkan bahwa pekerja tersebut bekerja pada akhir pekan, terutama pada hari minggu dari pukul 18.00-20.00

d. Dari data yang diberikan, terlihat bahwa pada hari Minggu terdapat beberapa jam malam dengan penggunaan CPU yang tinggi. Oleh karena itu, kita dapat menyimpulkan bahwa pada hari Minggu pekerja tersebut mulai bekerja pada komputer mereka pada malam hari.
