```python
import pandas as pd
import matplotlib.pyplot as plt

# Membaca data CSV
df = pd.read_csv('/content/drive/MyDrive/BI/penjualan.csv', sep='\t', header=None)
df.columns = ['2008-01-01 00:00:00', 'CABANG-039', '039-053', 'ITM-038',
       'PROD-0000040', '12']

# Mengganti tipe data kolom Tanggal menjadi datetime
df['2008-01-01 00:00:00'] = pd.to_datetime(df['2008-01-01 00:00:00'])

# Mengganti indeks menjadi kolom 'Produk'
df.set_index('PROD-0000040', inplace=True)

# Menghitung total penjualan untuk setiap produk
total_penjualan = df.groupby('PROD-0000040')['12'].sum().sort_values(ascending=False)

# Mengambil 10 produk teratas
top_10_produk = total_penjualan.head(10)

# Membuat grafik batang untuk menampilkan 10 produk teratas
top_10_produk.plot(kind='bar', figsize=(10, 6))
plt.title('Top 10 Produk Berdasarkan Total Penjualan')
plt.xlabel('Produk')
plt.ylabel('Total Penjualan')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Membaca data CSV
df = pd.read_csv('/content/drive/MyDrive/BI/harga_harian.csv', sep='\t', header=None)
df.columns = ['PROD-0000001', '2008-01-01 00:00:00', 'CABANG-039', '10980', '10380', '190']

# Mengganti tipe data kolom Tanggal menjadi datetime
df['2008-01-01 00:00:00'] = pd.to_datetime(df['2008-01-01 00:00:00'])

# Melakukan plotting menggunakan grafik batang
fig, ax = plt.subplots()

# Misalnya kita ingin memplot harga berlaku, modal, dan biaya untuk setiap cabang
cabang_labels = df['CABANG-039'].unique()
width = 0.2

# Menentukan posisi batang secara manual
positions_harga = range(len(cabang_labels))
positions_modal = [pos + width for pos in positions_harga]
positions_biaya = [pos + 2 * width for pos in positions_harga]

# Memplot bar untuk setiap kolom
ax.bar(positions_harga, df['10980'].values[:len(cabang_labels)], width=width, label='Harga Berlaku')
ax.bar(positions_modal, df['10380'].values[:len(cabang_labels)], width=width, label='Modal Cabang')
ax.bar(positions_biaya, df['190'].values[:len(cabang_labels)], width=width, label='Biaya Cabang')

ax.set_xticks([pos + width for pos in positions_harga])
ax.set_xticklabels(cabang_labels)
ax.set_xlabel('Cabang')
ax.set_ylabel('Jumlah')
ax.set_title('Perbandingan Harga Berlaku, Modal, dan Biaya per Cabang')
ax.legend()

# Menampilkan grafik
plt.show()
```

1. **Insight Grafik:**
   Grafik memberikan pemirsa wawasan tentang perbandingan harga berlaku, modal cabang, dan biaya cabang untuk setiap cabang (CABANG-039). Setiap batang dalam grafik mewakili satu cabang, dan tinggi batang mencerminkan nilai dari masing-masing parameter (harga berlaku, modal cabang, biaya cabang). Dengan demikian, kita dapat melihat perbandingan kinerja finansial antar cabang dengan mudah. Pada kasus ini membahas CABANG-039 secara menyeluruh

2. **Alasan Pemilihan Grafik:**
   - **Kejelasan Representasi:** Grafik batang sangat efektif dalam merepresentasikan perbandingan antar kategori, dalam hal ini, perbandingan parameter finansial antar cabang.
   - **Keterbacaan dan Kesederhanaan:** Grafik ini memberikan visualisasi yang jelas dan mudah dimengerti, dengan setiap cabang diidentifikasi dan perbandingan antar parameter yang dapat dilihat dengan cepat.
   - **Pemisahan Kategori:** Penggunaan warna atau label yang berbeda pada setiap batang memudahkan pemisahan dan identifikasi setiap kategori.

3. **Saran Strategi yang Dapat Ditempuh:**
   - **Optimasi Harga Berlaku:** Cabang dengan harga berlaku rendah dapat dievaluasi lebih lanjut untuk mengidentifikasi faktor-faktor yang mempengaruhi dan untuk merancang strategi peningkatan harga.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Membaca data CSV
df = pd.read_csv('/content/drive/MyDrive/BI/penjualan.csv', sep='\t', header=None)
df.columns = ['2008-01-01 00:00:00', 'CABANG-039', '039-053', 'ITM-038',
              'PROD-0000040', '12']

# Mengganti tipe data kolom Tanggal menjadi datetime
df['2008-01-01 00:00:00'] = pd.to_datetime(df['2008-01-01 00:00:00'])

# Menghitung total penjualan untuk setiap item
total_penjualan = df.groupby('ITM-038')['12'].sum().sort_values(ascending=False)

# Mengambil 10 item teratas
top_10_item = total_penjualan.head(10)

# Menghitung persentase penjualan
persentase_penjualan = (top_10_item / top_10_item.sum()) * 100

# Membuat diagram lingkaran untuk menampilkan proporsi penjualan item
plt.figure(figsize=(8, 8))
plt.pie(persentase_penjualan, labels=top_10_item.index, autopct='%1.1f%%', startangle=140)
plt.title('Proporsi Penjualan 10 Item Teratas')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Membaca data CSV
df = pd.read_csv('/content/drive/MyDrive/BI/penjualan.csv', sep='\t', header=None)
df.columns = ['2008-01-01 00:00:00', 'CABANG-039', '039-053', 'ITM-038', 'PROD-0000040', '12']


# Mengganti tipe data kolom Tanggal menjadi datetime
df['2008-01-01 00:00:00'] = pd.to_datetime(df['2008-01-01 00:00:00'])

# Mengelompokkan data berdasarkan Tanggal dan Kode Item
penjualan_per_hari = df.groupby(['2008-01-01 00:00:00', 'ITM-038'])['12'].sum().reset_index()

# Membuat tabel pivot untuk mendapatkan format yang sesuai untuk plot
pivot_table = penjualan_per_hari.pivot(index='2008-01-01 00:00:00', columns='ITM-038', values='12')

# Menampilkan grafik garis untuk menampilkan produk yang laku per hari
pivot_table.plot(kind='line', marker='o', figsize=(12, 8))
plt.title('Penjualan Produk per Hari dengan Kode Item')
plt.xlabel('Tanggal')
plt.ylabel('Total Penjualan')
plt.legend(title='Kode Item', loc='upper left', bbox_to_anchor=(1, 1))
plt.show()

```

1. **Insight dari Grafik:**
   - **Tren Penjualan Harian:** Grafik garis memungkinkan kita melihat tren penjualan harian untuk masing-masing item secara visual. Apakah ada peningkatan atau penurunan dalam penjualan suatu item dari waktu ke waktu dapat terlihat dengan jelas.
   - **Perbandingan Antara Item:** Dengan menggunakan grafik garis, kita dapat dengan mudah membandingkan kinerja penjualan harian antara item-item berbeda. Ini membantu dalam mengidentifikasi item yang mungkin lebih populer atau mengalami pertumbuhan yang lebih cepat.

2. **Alasan Pemilihan Grafik Garis:**
   - **Representasi Tren:** Grafik garis secara alami cocok untuk merepresentasikan tren dan perubahan sepanjang waktu. Ini memberikan visibilitas yang baik terhadap bagaimana pola penjualan berkembang dari hari ke hari.
   - **Bandingkan Nilai:** Dengan menggunakan garis terpisah untuk setiap item, kita dapat dengan mudah membandingkan nilai penjualan harian mereka. Ini memberikan pemahaman yang lebih baik tentang relatifnya kinerja item satu terhadap yang lain.

3. **Saran Strategi:**
   - **Promosi Berdasarkan Tren:** Jika ada item dengan tren penjualan yang kuat, perusahaan dapat merencanakan promosi atau strategi pemasaran yang lebih fokus pada item tersebut.
   - **Manajemen Stok:** Dengan memahami tren penjualan harian, perusahaan dapat mengelola stok mereka dengan lebih efisien, memastikan ketersediaan produk saat dibutuhkan dan menghindari kelebihan stok.
   - **Penyesuaian Harga:** Jika ada fluktuasi yang signifikan dalam penjualan harian, penyesuaian harga taktis dapat dipertimbangkan untuk meningkatkan daya tarik pelanggan.
   - **Analisis Item Bersamaan:** Perhatikan apakah ada pola penjualan bersamaan antara beberapa item. Ini dapat memberikan wawasan tentang item-item yang sering dibeli bersamaan, yang dapat mengarah pada strategi bundel atau penawaran paket.
